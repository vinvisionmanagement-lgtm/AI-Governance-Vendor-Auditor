```python
import spacy

# Load spaCy model (do this once at the beginning of your script)
nlp = spacy.load("en_core_web_md")

# Convert text to lowercase so we don't miss "BAA" vs "baa"
clean_text = vendor_text.lower()
# Process the text with spaCy
doc = nlp(clean_text)

print(f"{'='*10} INITIAL AI AUDIT REPORT {'='*10}\n")

# Logic to find HIPAA blockers with similarity
for term in hipaa_terms:
    # Exact match check
    if term.lower() in clean_text:
        print(f"[+] FOUND COMPLIANCE TERM: {term}")
    else:
        # Similarity check
        term_doc = nlp(term.lower())
        # Check similarity with chunks of text
        found = False
        for chunk in doc.noun_chunks:
            similarity = term_doc.similarity(chunk)
            if similarity > 0.8:  # Threshold can be adjusted
                print(f"[+] FOUND SIMILAR COMPLIANCE TERM: '{chunk}' (similarity: {similarity:.2f}) to '{term}'")
                found = True
                break
        
        # This is a specific 'Lawyer' concern: Missing terms
        if term == "BAA" and not found:
            print(f"[!] WARNING: Missing {term}. This tool may not be HIPAA compliant.")

# Logic to find Security/AI Risks with similarity
for term, risk_level in ai_risk_terms.items():
    # Exact match check
    if term.lower() in clean_text:
        print(f"[{risk_level} RISK]: This vendor uses the phrase '{term}'.")
    else:
        # Similarity check
        term_doc = nlp(term.lower())
        for chunk in doc.noun_chunks:
            similarity = term_doc.similarity(chunk)
            if similarity > 0.8:  # Threshold can be adjusted
                print(f"[{risk_level} RISK]: This vendor uses a phrase similar to '{term}': '{chunk}' (similarity: {similarity:.2f})")
                break
```

```python
import datetime

# Create a list to hold our final report data
audit_report = []

for term, risk_level in ai_risk_terms.items():
    if term.lower() in vendor_text.lower():
        # Build a record for this finding
        finding = {
            "Term Found": term,
            "Risk Level": risk_level,
            "Recommendation": mitigation_strategies.get(risk_level, "N/A"),
            "Date Flagged": datetime.date.today()
        }
        audit_report.append(finding)

# Print a summary to the console for a quick check
print(f"Audit Complete. Found {len(audit_report)} flagged terms.")
```

```python
import csv

# Define the columns for your spreadsheet
fieldnames = ["Term Found", "Risk Level", "Recommendation", "Date Flagged"]

# Write to a file
with open('ai_vendor_risk_report.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(audit_report)

print("Report exported to: ai_vendor_risk_report.csv")
```
